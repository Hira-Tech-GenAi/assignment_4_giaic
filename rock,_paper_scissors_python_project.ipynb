## Code Explanation
The Python code is simple and beginner-friendly. Here's a breakdown:

- **Imports**: The `random` module is used to generate the computer's choice.
- **Function**: The `play_game()` function contains the game logic.
- **Choices**: A dictionary maps user inputs (`1`, `2`, `3`) to choices with emojis (`‚úä Rock`, `üìÑ Paper`, `‚úÇÔ∏è Scissors`).
- **Game Loop**:
  - Prompts the user for input and validates it.
  - Randomly selects the computer's choice.
  - Displays both choices and determines the winner based on the rules.
  - Asks if the user wants to play again.
- **Emojis**: Used to represent choices and enhance the user experience (e.g., üèÜ for winning, ü§ù for a tie).

```python
import random

def play_game():
    print("üéÆ Welcome to Rock, Paper, Scissors! üéâ")
    print("Winning rules: ‚úä Rock smashes ‚úÇÔ∏è Scissors, üìÑ Paper covers ‚úä Rock, ‚úÇÔ∏è Scissors cuts üìÑ Paper")

    choices = {"1": "‚úä Rock", "2": "üìÑ Paper", "3": "‚úÇÔ∏è Scissors"}

    while True:
        print("\nEnter your choice:")
        print("1 - ‚úä Rock")
        print("2 - üìÑ Paper")
        print("3 - ‚úÇÔ∏è Scissors")

        user_choice = input("Your choice (1-3): ").strip()

        if user_choice not in ["1", "2", "3"]:
            print("‚ùå Invalid choice! Please enter 1, 2, or 3.")
            continue

        user_action = choices[user_choice]
        computer_action = random.choice(list(choices.values()))

        print(f"\nYou chose: {user_action}")
        print(f"Computer chose: {computer_action}")

        if user_action == computer_action:
            print("ü§ù It's a tie!")
        elif (user_action == "‚úä Rock" and computer_action == "‚úÇÔ∏è Scissors") or \
             (user_action == "üìÑ Paper" and computer_action == "‚úä Rock") or \
             (user_action == "‚úÇÔ∏è Scissors" and computer_action == "üìÑ Paper"):
            print("üèÜ You win!")
        else:
            print("üòû Computer wins!")

        play_again = input("\nPlay again? (y/n): ").lower()
        if play_again != "y":
            print("üëã Thanks for playing!")


play_game()
```
