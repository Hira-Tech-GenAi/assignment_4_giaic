# üß† Guess the Number (Computer Guesses) üß†

## Project Overview
Welcome to the **Reverse Guess the Number Game**!  
In this project, **you think of a number between 1 and 50**, and the **computer will try to guess it** using your feedback.

---

### üéØ Features:
- You think of a number between **1 and 50** üß†
- The computer makes intelligent guesses using **binary search logic** üìà
- You give feedback:  
  - `h` = too high üî∫  
  - `l` = too low üîª  
  - `c` = correct ‚úÖ
- Fast and efficient guessing üéØ
- Fun and interactive with emoji feedback üéâ

---

## üõ†Ô∏è How It Works:
1. The user thinks of a number between 1 and 50.
2. The computer guesses the middle number between the current low and high range.
3. The user gives feedback on each guess:
   - `h` if the guess is too **high**
   - `l` if the guess is too **low**
   - `c` if the guess is **correct**
4. The computer adjusts its guess based on the feedback until it finds the correct number.

```python
def computer_guess():
    print("ü§ñ Welcome to the Computer Guess Game!")
    print("üß† Think of a number between 1 and 50, and I'll try to guess it.")
    input("üéØ Press Enter when you're ready...")

    low = 1
    high = 50
    attempts = 6

    while low <= high:
        guess = (low + high) // 2
        attempts += 1
        print(f"\nü§î My guess is: {guess}")
        feedback = input("Is it (h)igh, (l)ow, or (c)orrect? ").lower()

        if feedback == 'c':
            print(f"üéâ Yay! I guessed your number in {attempts} attempts!")
            break
        elif feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1
        else:
            print("‚ö†Ô∏è Please enter a valid response: 'h', 'l', or 'c'.")

    if low > high:
        print("üòï Hmm, something went wrong. Were you honest about your number?")

computer_guess()
```
