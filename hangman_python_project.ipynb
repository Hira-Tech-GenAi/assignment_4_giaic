# üéâ Hangman Game

Welcome to the Hangman Game, a classic and interactive Python-based word game! This project challenges you to guess a fruit word letter by letter, guided by a hint, before running out of lives.

---

## üß† What is Hangman?

Hangman is a word-guessing game where you try to figure out a hidden word by guessing one letter at a time. Each correct guess reveals letters in the word, but each wrong guess brings you closer to losing the game. In this version, all words are fruits, and you get a helpful hint to start!

---

## üöÄ Features

- üçé **Fruit-themed words** with descriptive hints  
- ‚ù§Ô∏è **6 lives** to guess the word  
- üéâ **Emojis** for a fun, visual experience  
- üìù **Detailed comments** for easy code understanding  
- üíª **Beginner-friendly** Python code

```python
import random  # Import random module to select a word randomly

def hangman():
    # Dictionary mapping each fruit word to a descriptive hint
    word_hints = {
        'apple': 'This fruit is often red or green and used in pies üçé',
        'banana': 'This fruit is yellow and curved üçå',
        'mango': 'This fruit is juicy and orange inside ü•≠',
        'strawberry': 'This fruit is red with tiny seeds on the outside üçì'
    }
    # Randomly choose a word from the dictionary keys
    word = random.choice(list(word_hints.keys()))
    # Get the corresponding hint for the chosen word
    hint = word_hints[word]
    # Convert word to a set of unique letters to track guessed letters
    word_letters = set(word)
    # Create a set of all lowercase alphabet letters for validation
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    # Initialize an empty set to store letters guessed by the player
    used_letters = set()
    # Set the number of lives (incorrect guesses allowed)
    lives = 6

    # Display the hint at the start to help the player
    print(f'Hint: {hint} ‚ùì')

    # Loop until all letters are guessed or lives run out
    while len(word_letters) > 0 and lives > 0:
        # Show remaining lives and letters already used
        print(f'You have {lives} lives ‚ù§Ô∏è left. Used letters: {" ".join(used_letters)} üìù')
        # Create a list showing guessed letters and dashes for unguessed ones
        word_list = [letter if letter in used_letters else '-' for letter in word]
        # Display the current state of the word
        print(f'Word: {" ".join(word_list)} üîç')

        # Prompt player to guess a letter (simple English, indicating it's a fruit)
        user_letter = input('Guess a letter, it\'s a fruit: ').lower()
        # Check if the input is a new, valid letter
        if user_letter in alphabet - used_letters:
            # Add the guessed letter to used letters
            used_letters.add(user_letter)
            # If the letter is in the word, remove it from word_letters
            if user_letter in word_letters:
                word_letters.remove(user_letter)
                # Confirm correct guess
                print(f'Correct! {user_letter} is in the word üéâ')
            else:
                # Reduce lives for incorrect guess
                lives -= 1
                # Inform player of incorrect guess
                print(f'Wrong! {user_letter} not in word üíî')
        # Handle case where letter was already guessed
        elif user_letter in used_letters:
            print(f'Already guessed {user_letter} üîÑ. Try again!')
        # Handle invalid input (e.g., numbers or symbols)
        else:
            print('Invalid input üö´. Enter a letter.')

    # Check game outcome after loop ends
    if lives == 0:
        # Display loss message with the correct word
        print(f'Game over üò¢. The word was {word}')
    else:
        # Display win message with the correct word
        print(f'You win! The word was {word} üèÜ')

# Start the game
hangman()
```
