```python
import time  # Import time module for countdown

# Function to start the countdown
def start_countdown(minutes, seconds):
    # Convert total time to seconds
    total_seconds = minutes * 60 + seconds

    try:
        while total_seconds >= 0:
            # Use divmod to get minutes and seconds
            mins_display, secs_display = divmod(total_seconds, 60)

            # Print countdown in MM:SS format
            print(f"\r‚è≥ Time Left: {mins_display:02d}:{secs_display:02d}", end="")

            time.sleep(1)  # Wait for 1 second
            total_seconds -= 1  # Decrease time

        print("\n‚è∞ Time's up!")  # Message when countdown finishes

    except KeyboardInterrupt:
        print("\n‚ùó Countdown stopped manually.")

# Main function to get input from user
def main():
    print("üéØ Welcome to the Countdown Timer!")

    try:
        # Get minutes and seconds input from user
        mins = int(input("Enter minutes: "))
        secs = int(input("Enter seconds: "))

        # Validate input
        if mins < 0 or secs < 0 or secs >= 60:
            print("‚ö†Ô∏è Please enter valid minutes (>=0) and seconds (0-59).")
            return

        # Start the countdown
        start_countdown(mins, secs)

    except ValueError:
        print("‚ùó Invalid input! Please enter numbers only.")

# Run the program
if __name__ == "__main__":
    main()
```
