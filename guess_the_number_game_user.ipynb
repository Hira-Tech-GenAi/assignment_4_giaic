# üéÆ Guess the Number Game üéÆ

## Project Overview
Welcome to the **Guess the Number Game**! In this project, you'll guess a number between 1 and 50. You have 5 attempts to guess it correctly.

### Features:
- Randomly generated number between 1 and 50 üé≤
- 5 attempts to guess the number üß†
- Hints on whether your guess is too high or too low üìâüìà
- Fun and interactive with emojis üéâ

---

## How It Works:
1. The game generates a random number between 1 and 50.
2. The user has 5 attempts to guess the number.
3. After each guess, the program will tell the user if the guess is:
   - Too low ‚¨áÔ∏è
   - Too high ‚¨ÜÔ∏è
   - Correct üéâ
4. The user must guess correctly within 5 attempts or the game will reveal the correct number üòû

```python
import random

def guess_number():
    # Generate a random number between 1 and 50
    number_to_guess = random.randint(1, 50)
    attempts = 5  # Set the number of attempts to 5

    print("üéâ Welcome to the Guess the Number Game! üéâ")
    print("ü§î I'm thinking of a number between 1 and 50. ü§î")
    print("üéØ You have 5 attempts to guess it! üéØ")

    while attempts > 0:
        try:
            # Get user input
            user_guess = int(input(f"\nYou have {attempts} attempts remaining. Enter your guess: "))

            # Check if the guess is within the valid range
            if user_guess < 1 or user_guess > 50:
                print("‚ö†Ô∏è Please guess a number between 1 and 50. ‚ö†Ô∏è")
                continue

            # Check if the guess is correct
            if user_guess == number_to_guess:
                print(f"üéâ Congratulations! You guessed the number {number_to_guess} correctly! üéâ")
                break
            elif user_guess < number_to_guess:
                print("‚¨áÔ∏è Too low! Try again. ‚¨áÔ∏è")
            else:
                print("‚¨ÜÔ∏è Too high! Try again. ‚¨ÜÔ∏è")

            # Decrease the number of attempts
            attempts -= 1

        except ValueError:
            print("‚ùå Invalid input! Please enter a valid number. ‚ùå")

    if attempts == 0:
        print(f"\nSorry! You've used all your attempts. The correct number was {number_to_guess}. üòû")

# Call the function to start the game
guess_number()
```
